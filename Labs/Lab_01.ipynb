```python
# LAB 1: BETA DISTRIBUTIONS

# number of draws
N = 10000

# set shape parameters
betaParams1 = [0.5,0.5]
betaParams2 = [2,2]
betaParams3 = [2,5]
betaParams4 = [1,1]

# sample from the distributions N times
betaDist1 = np.random.beta(betaParams1[0],betaParams1[1],N)
betaDist2 = np.random.beta(betaParams2[0],betaParams2[1],N)
betaDist3 = np.random.beta(betaParams3[0],betaParams3[1],N)
betaDist4 = np.random.beta(betaParams4[0],betaParams4[1],N)
```

```python
plt.hist(betaDist1,25)
plt.xlabel("x")
plt.ylabel("counts")
```

```python
# HOMEWORK-LIKE EXAMPLE

# Exponential distribution parameters
lambdaMeeting = 1/2.
lambdaLunch = 3.

# But note the correct parametrization!
betaMeeting = 1/lambdaMeeting
betaLunch = 1/lambdaLunch

meetingMax = 1.5
lunchMin = .5

N = 100000
meetingDist = np.random.exponential(betaMeeting,N)
lunchDist = np.random.exponential(betaLunch,N)

# probability of productive day = P(lunch > lunchMin)*P(meeting < meetingMax)

    # Note: this is probably not the nicest way to do it programmatically in Python - see if you can do better! 
probProductiveDay = 1.0*sum(1*(meetingDist<meetingMax))/N*sum(1*(lunchDist > lunchMin))/N
                                                              
print "The probability of a productive day is:", probProductiveDay
```
